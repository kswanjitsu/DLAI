```python
acab_list = ['(R)', '10L', '11L', '12L', '1L', '2L', '3L', '4L', '5L', '6L', '7L', '8L', '9L', 'A.D.L.', 'A.S.A.P.', 'A.T.', 'A1c', 'A:', 'AAOx1', 'AAOx2', 'AAOx3', 'AAOx4', 'AAO', 'AA', 'ABG', 'ABI', 'ABP', 'AB', 'ACE', 'ACLS', 'AC', 'ADA', 'ADL', 'AD', 'AF', 'ALD', 'ALT', 'AMA', 'AM', 'ANA', 'APC', 'ASAP', 'ASA', 'AVR', 'AV', 'AWOL', 'Aa', 'Abd', 'Ab', 'Adm', 'Alc', 'Alt', 'Amt', 'Ant', 'Ax', 'B/P', 'BAL', 'BE', 'BID', 'BKA', 'BK', 'BMI', 'BMP', 'BMT', 'BM', 'BNP', 'BP', 'BRBPR', 'BS', 'BUN', 'BiPAP', 'Bx', 'C&S', 'C3', 'C4', 'CABG', 'CAD', 'CA', 'CBC', 'CC', 'CD4', 'CDI', 'CEA', 'CK', 'CL', 'CNS', 'CO2', 'COPD', 'CPA', 'CPK', 'CPR', 'CP', 'CRP', 'CRRT', 'CR', 'CSF', 'CTAB', 'CTA', 'CT', 'CVA', 'CVP', 'CVS', 'Ca++', 'CaS', 'Chem7', 'Cl', 'D&C', 'D.O', 'D/C', 'D5W', 'DC summary', 'DC', 'DDx', 'DD', 'DE', 'DIP', 'DM1', 'DM2', 'DM', 'DNA', 'DOA', 'DOB', 'DOE', 'DO', 'DP', "DT's", 'DTRs', 'DTR', 'DT', 'DVT', 'Derm', 'Diff', 'Dx', 'ECG', 'ECT', 'EC', 'EEG', 'EENT', 'EKG', 'ENT', 'EOM', 'EP', 'ERCP', 'ER', 'ESR', 'EST', 'ES', 'ETT', 'ET', 'Elev', 'EtOH', 'Ext', 'FBS', 'FISH', 'FSH', 'FUO', 'Fa', 'Fl', 'Fx', 'GB', 'GGT', 'GI', 'GSW', 'GTT', 'GT', 'GU', 'Gen', 'Gp', 'Gtts', 'Gyn', 'H202', 'H20', 'HA', 'HCVD', 'HD', 'HEENT', 'HHFNC', 'HPI', 'HR', 'HS', 'HbA1c', 'Hbg', 'Hb', 'Hct', 'I&D', 'I&O', 'IA', 'IB', 'IC', 'IJ', 'IM', 'IPPB', 'IR', 'ITP', 'IT', 'IVC', 'IVF', 'IVP', 'IV', 'K+', 'KUB', 'K', 'L&A', 'LAD', 'LA', 'LE', 'LFT', 'LLQ', 'LOA', 'LPN', 'LP', 'LR', 'LUQ', 'LV', 'Leuc', 'Lg', 'L', 'M.D.', 'M.T.', 'MCHC', 'MCH', 'MCV', 'MD', 'MI', 'MOM', 'MP', 'MR#', 'MRCP', 'MRI', 'MR', 'MSK', 'MSSA', 'MS', 'Mg', 'Mm', 'Mo', 'NAD', 'NA', 'NBNB', 'NBS', 'NCAT', 'NC', 'NKA', 'NKDA', 'NMR', 'NPN', 'NPO', 'NP', 'NSS', 'Na+', 'Na', 'Neg', 'Neuro', 'Ng', 'Noct', 'O.T.', 'O2', 'O:', 'OBS', 'OB', 'OD', 'OOB', 'OP', 'OR', 'OS', 'OTC', 'OT', 'OU', 'Od', 'Opd', 'Op', 'Ortho', 'Oz', 'P&A', 'P.E.', 'P.I.D.', 'P:', 'PAC', 'PA', 'PBI', 'PCA', 'PCP', 'PDA', 'PD', 'PEG', 'PERLA', 'PERRLA', 'PERRL', 'PE', 'PMHx', 'PMR', 'PM', 'PN', 'PO', 'PPD', 'PRN', 'PR', 'PSA', 'PSHx', 'PSH', 'PTT', 'PT', 'Pc', 'Pdr', 'Peds', 'Pen', 'Per', 'PhD', 'Psy Ser', 'P', 'QAC', 'QAP', 'QAR', 'QA', 'QHS', 'R.D.', 'R/O', 'RA', 'RBC', 'RD', 'RLQ', 'RN', 'ROS', 'RR', 'RT', 'RUQ', 'RVR', 'Rh', 'Rx', 'R', 'S/A', 'S2', 'S4', 'S:', 'SA', 'SBP', 'SGOT', 'SGPT', 'SIDS', 'SMA', 'SOAP', 'SOB', 'SQ', 'SSD', 'SS', 'SW', 'Sa', 'Schiz', 'Sibs', 'Sig', 'Soc Ser', 'Soc hx', 'Sol', 'Sos', 'Spec', 'Sp', 'Staph', 'Stat', 'Strep', 'Sup', 'Surg', 'Sx', 'Sys', 'T & A', 'T.A.', 'T.O.', 'T1', 'T2', 'T3', 'T4', 'TB', 'TIA', 'TID', 'TLC', 'TM', 'TPA', 'TPRT', 'TP', 'TTP', 'Tab', 'Tbsp', 'Tmax', 'Tx', 'UA', 'UGIB', 'UGI', 'UNK', 'UO', 'URI', 'US', 'UTI', 'V.O.', 'VAD', 'VATS', 'VBG', 'VC', 'VDRL', 'VD', 'VSS', 'VS', 'VTE', 'Vag', 'WBC', 'WNL', 'a fib', 'a.c.', 'a/c', 'abd', 'ad lib', 'afib', 'amb', 'angio', 'ante', 'approx', 'appt', 'aq', 'art', 'as tol', 'atach', 'b.i.d.', 'blad', 'c/o', 'c/s', 'cal', 'cap', 'cc', 'chol', 'clav', 'cm', 'comf', 'conc', 'cons', 'contrib', 'cont', 'cor', 'cpd', 'cran', 'd/c', 'd5', 'decr', 'defib', 'depr', 'dil', 'dist', 'dis', 'dk', 'dl', 'dopa', 'etx', 'extr', 'f/c', 'f/u', 'ft', 'fu', 'genit', 'gluc', 'gr.', 'grav.', 'gtt', 'h/o', 'hern', 'hr', 'ht', 'hx', 'hyst', 'id', 'inj', 'inpt', 'int', 'irreg', 'irrig', 'kg', 'labs', 'lab', 'lac', 'lap', 'lat', 'lb', 'lig', 'liq', 'lymphs', 'm.s.', 'mEq', 'mae', 'max', 'meds', 'med', 'melena', 'mg', 'norm', 'nullip', 'od', 'outpt', 'p.o.', 'pH', 'palp', 'phys', 'picc', 'pmh', 'pneu', 'postop', 'post', 'pos', 'preop', 'prep', 'prev', 'prn', 'pros', 'prox', 'pulm', 'pvc', 'pwd', 'q.d.', 'q.h.', 'q.i.d.', 'q.o.d.', 'qh', 'qns', 'qod', 'q', 'r/o', 'rads', 'rbc', "rec'd", 'rect', 'ro', 'rpt', 's/p', 'sed rate', 'serol', 'sp gr', 'sp. cd.', 'sp', 'subcut', 'subq', 't.i.d.', 'temp', 'tfx', 'thor', 'thromb', 'trx', 'tsp', 'u/o', 'vent', 'vfib', 'vit', 'vol', 'vtach', 'wbc', 'wd', 'wf', 'wk', 'wnl', 'wn', 'wt', 'x', 'y/o', 'yo', 'yrs']

def create_output(acab):
    print(acab)
    acab_count = 0
    acab = acab.strip('\n').lower()
    for line in open('./datasets/dataset_parsed_all_mimic_test_notes.txt'):
        doc = nlp(line)
        for token in doc:
            #print(i)
            if acab in token.text.lower():
                acab_count += 1
    acab_result_list = [acab, acab_count]
    return acab_result_list
```

```python
delayed_results = [delayed(create_output)(acab) for acab in acab_list]
final_results = compute(*delayed_results, scheduler="threads")
```
